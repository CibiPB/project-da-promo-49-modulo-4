### Columna de tipo de crimen 🪒

```python
crime_keywords = {

    'Pedophilia': ['Child','child-murderer','pedophilia', 'child abuse', 'child molestation', 'child rape', 'sexual assault on children', 'sexual exploitation of children'],
    'Kidnapping': ['kidnap', 'abduction'],
    'Rape': ['rape','rapist-killer', 'sexual assault', 'molestation'],
    'Torture': ['torture', 'mutilation', 'disembowelment'],
    'Robbery': ['robbery', 'theft', 'steal'],
    'Poisoning': ['poison', 'toxins', 'overdose'],
    'Arson': ['arson', 'fire', 'burn'],
    'Necrophilia': ['necrophilia', 'corpse abuse'],
    'Cannibalism': ['canibalized','cannibalism', 'eating', 'consumed'],
    'Angel of Mercy': ['angel of mercy', 'angel of death', 'caregiver', 'nurse'],
    'Murder':["killing"]
}
def categorize_crime(notes, crime_keywords):
    if isinstance(notes, str):  # Verificar si 'notes' es un string
        for crime, keywords in crime_keywords.items():
            if any(keyword in notes.lower() for keyword in keywords):  # Convertir a minúsculas
                return crime
    return 'Other'

# Aplicar la función al DataFrame
df_killers_byvictims['Crime Type'] = df_killers_byvictims['Notes'].apply(categorize_crime, crime_keywords=crime_keywords)

```
